```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('dark_background')
```

## matplotlib with numpy

```python
x=np.arange(1,11) # used to create an array whose value is in range of 1 to 10
y=np.arange(10,110,10)   # in this value ranges from 10 to 100 with difference of 10 (as we have step == 10 )
plt.figure(figsize=(6,6))
plt.plot(x,y,"r--")

plt.show()
```

```python
x = np.arange(0, 2*np.pi, 0.1)
y=np.sin(x)
plt.figure(figsize=(6,6))
plt.plot(x,y,"r--")
plt.xlabel("Angle (radians)")
plt.ylabel("Sine")
plt.title("Sine Graph")
plt.show()
```

```python
x=np.arange(-1,1,.1)
y=abs(x)
plt.plot(x,y,"-r")
plt.show()
```

```python
x=np.arange(0,2*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y,"-r")
plt.show()
```

## Playing numpy with strings


```python
s1= "hello everyone this is google collab and "
s2= "soon i am going to learn django framework"
```

```python
np.char.add(s1,s2) , np.char.upper(s1)
```

```python
# to convert to lower case use np.char.lower

# to split on a delimiter use np.char.split

np.char.splitlines(s1) # split the string when eof found
np.char.split(s1)
```

```python
np.char.replace(s1,"google", "jupyter") # to replace something from a string
```

```python
# To print something on center
np.char.center("text", 100, "-")
```

## panda series with python list

```python
l1=np.arange(1,10,1) # creating an array from 1 to 9
# creating a panda series from array
series=pd.Series(l1)
print(series)   # printing series creates a table with index starting from 0
```

```python
# empty Series
empty= pd.Series([])
print(empty)
```

```python
# Creating your own series index
Myown_series=pd.Series(['p','i','y','u','s','h'], index=[10,11,12,13,14,15])
Myown_series
```

```python
# Giving name to the series
Myown_series=pd.Series(['p','i','y','u','s','h'], index=[10,11,12,13,14,15], name="Alias")
Myown_series
```

```python
# panda series with scalar values give one value only
pd.Series(2,index=[1,2,3,4,5,6])
```
